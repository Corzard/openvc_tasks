Найти все orb features точки на изображении

```python
img = cv.imread('opencv.png',0)
# Initiate ORB detector
orb = cv.ORB_create()
# find the keypoints with ORB
kp = orb.detect(img,None)
# compute the descriptors with ORB
kp, des = orb.compute(img, kp)
# draw only keypoints location,not size and orientation
img2 = cv.drawKeypoints(img, kp, None, color=(0,255,0), flags=0)
plt.imshow(img2), plt.show()
```

Найти все sift features точки на изображении

```python
img = cv.imread('opencv.png')
gray= cv.cvtColor(img,cv.COLOR_BGR2GRAY)
sift = cv.SIFT_create()
kp = sift.detect(gray,None)
img3=cv.drawKeypoints(gray,kp,img)
plt.imshow(img3), plt.show()
```

```python
import numpy as np
import cv2 as cv
cap = cv.VideoCapture(0)
if not cap.isOpened():
    print("Cannot open camera")
    exit()
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    # Our operations on the frame come here
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    sift = cv.SIFT_create()
    kp = sift.detect(gray,None)
    capimg=cv.drawKeypoints(gray,kp,frame)
    # Display the resulting frame
    cv.imshow('frame', capimg)
    if cv.waitKey(1) == ord('q'):
        break
# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
```

Повернуть изображение на 45 градусов

```python
(h, w) = img.shape[:2]
(cX, cY) = (w // 2, h // 2)
M = cv.getRotationMatrix2D((cX, cY), 45, 1.0)
rotated = cv.warpAffine(img, M, (w, h))
plt.imshow(rotated), plt.show()
```

Изменить яркость изображения

```python
matrix = np.ones(img.shape, dtype = "uint8") * 100

img_rgb_brighter = cv.add(img, matrix)
plt.figure(figsize=[18,5])

plt.subplot(132), plt.imshow(img),       plt.title("Original")
plt.subplot(133), plt.imshow(img_rgb_brighter), plt.title("Brighter")
```

Изменить контрасть изображения

```python
matrix1 = np.ones(img.shape) * .8
matrix2 = np.ones(img.shape) * 1.2

img_rgb_darker   = np.uint8(cv.multiply(np.float64(img), matrix1))
img_rgb_brighter = np.uint8(cv.multiply(np.float64(img), matrix2))

# Show the images
plt.figure(figsize=[18,5])
plt.subplot(131), plt.imshow(img_rgb_darker),  plt.title("Lower Contrast")
plt.subplot(132), plt.imshow(img),  plt.title("Original")
plt.subplot(133), plt.imshow(img_rgb_brighter), plt.title("Higher Contrast")
```

Сделать гамма-перобразование 

Сделать гистограмную эквайлизацию 

```python
im = cv.imread("env.jpg")

# split g,b,r
g = im[:,:,0]
b = im[:,:,1]
r = im[:,:,2]


# Histogram Equalization
r2 = cv.equalizeHist(r)
g2 = cv.equalizeHist(g)
b2 = cv.equalizeHist(b)

im2 = im.copy()
im2[:,:,0] = g2
im2[:,:,1] = b2
im2[:,:,2] = r2

# print(im2)
plt.figure(figsize=[18,5])
plt.subplot(131),plt.imshow(im),  plt.title("Before")
plt.subplot(132),plt.imshow(im2),  plt.title("After")
plt.show()
```

Сделать бинаризацию изображения

```python
img = cv.imread('env2.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray, cmap ='gray'), plt.show()
ret, th = cv.threshold(gray, 100,255, cv.THRESH_BINARY)
plt.imshow(th, cmap ='gray'), plt.show()
```

Найти контуры на бинаризированном изображении

```python
contours, hierarchy = cv.findContours(th, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
cv.drawContours( img, contours, -1, (255,0,0), 3, cv.LINE_AA, hierarchy, 1 )
plt.imshow(img, cmap="gray"), plt.show()
```

```python
cap = cv.VideoCapture(0)
if not cap.isOpened():
    print("Cannot open camera")
    exit()

# Capture frame-by-frame
ret, frame = cap.read()
# if frame is read correctly ret is True
if not ret:
    print("Can't receive frame (stream end?). Exiting ...")
else:
    # Our operations on the frame come here
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    # plt.imshow(gray, cmap ='gray'), plt.show()
    ret, th = cv.threshold(gray, 100,255, cv.THRESH_BINARY)
    plt.imshow(th), plt.show()
    contours, hierarchy = cv.findContours(th, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
    cv.drawContours(gray, contours, -1, (255,0,0), 2, cv.LINE_AA, hierarchy, 1 )
    plt.imshow(gray, cmap='gray'), plt.show()

# When everything done, release the capture
cap.release()
```
